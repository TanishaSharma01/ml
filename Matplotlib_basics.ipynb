**Matplotlib**

*   Useful for making plots



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy library
import numpy as np
```

```python
# gets evenly spaced 100 values between 0 and 10
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title, x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# red plus signs
plt.plot(x, y, 'r+')
plt.show()
```

```python
# green dots
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
# adds an axes object to the figure left, bottom, width, height
ax = fig.add_axes([0,0,1,1])
# features
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
# bar plot for these two (x,y)
ax.bar(languages,people)
plt.xlabel('Language')
plt.ylabel('Number of people')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
# people: data values, labels=languages: names shown for each slice, autopct='%1.1f%%'  displays percentages
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
# Generates 30 evenly spaced values between 0 and 10
x = np.linspace(0, 10, 30)
# Computes the sine of each value in x
y = np.sin(x)
# Computes the cosine of each value in x
z = np.cos(x)
# Creates a new figure (canvas) for plotting
fig2 = plt.figure()
# Adds axes that fill the entire figure
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap='Blues')
plt.show()
```
