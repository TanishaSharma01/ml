```python
import numpy as np
```

List vs Numpy - Time taken


```python
from time import process_time
```

Time taken by a list

```python
python_list = [i for i in range(10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

Time taken by numpy aray

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5

end_time = process_time()

print(end_time - start_time)
```

Initial Placeholders in Numpy array

```python
# create array with all zeros
# mention in brackets the shape of your array
x = np.zeros((4,5))
print(x)
```

```python
# create a numpy array of ones
y = np.ones((4,5))
print(y)
```

```python
# array of a particular value
z = np.full((3,2),5)
print(z)
```

```python
# create an identity matrix
a = np.eye(4)
print(a)
```

```python
# create a numpy array with random values
b = np.random.random((3,4))
print(b)
```

```python
# create a numpy array with random integers
c = np.random.randint(10,100,(3,5))
print(c)
```

```python
# array of evenly spaced values --> specifying the number of values required
d = np.linspace(10, 30, 6)
print(d)
```

```python
# array of evenly spaced values --> specifying the steps
e = np.arange(10,30,5)
print(e)
```

```python
# convert a list to numpy array
list2 = [10,20,30,40,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

Mathematical operations on an np array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

print(list1 + list2) #concatenates or joins two lists
```

```python
a = np.random.randint(0, 10, (3,3))
b = np.random.randint(10, 20, (3,3))
```

```python
print(a)
print(b)
```

```python
print(a+b) # summed up the values of the two arrays
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
# transpose 2
trans2 = array.T
print(trans2)
```

```python
# reshaping an array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
